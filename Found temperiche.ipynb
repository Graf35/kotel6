```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import MultiTaskLassoCV
from sklearn.linear_model import Ridge
```

```python
df = pd.read_csv('uuu.csv', ";")
df
```

```python
df.info()
```

```python
y = df.drop(['k6f6x.PV', 'k6t2_1.PV', 'k6l2.PV', 'k6f6.PV'], axis=1)
x = df.drop(['k6ps14_1.PV', 'k6t17.PV', 'k6t15.PV', 'k6t16.PV', 'k6p8.PV', 'k6p13.PV', 
             'k0p102_1.PV', 'k6t6.PV', 'k6t14.PV', 'k6t3.PV', "k6f6.PV"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state = 15)
```

```python
y
```

```python
lasso = MultiTaskLassoCV(cv=5, random_state=10).fit(X_train, y_train)
```

```python
lasso.score(X_test, y_test)
```

```python
lasso.get_params()
```

```python
clf = Ridge(alpha=1.0)
clf.fit(X_test, y_test)
```

```python
clf.score(X_test, y_test)
```
